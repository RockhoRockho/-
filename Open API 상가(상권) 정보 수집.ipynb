## 상권조회

* 지정 상권조회(storeZoneOne)

```python
from urllib.request import Request, urlopen
from urllib.parse import urlencode, quote_plus
import pandas as pd
import json

def baro_api(skey, res_id, encode):
    
    url = 'http://apis.data.go.kr/B553077/api/open/sdsc2/baroApi'
    stype = 'json'
    params = f'?resId={res_id}&ServiceKey={skey}&type={stype}&' + encode
    query = url + params
    # print(query)

    request = Request(query)
    request.get_method = lambda: 'GET'
    response_body = urlopen(request).read()
    # print(response_body)
    response_dict = json.loads(response_body.decode('utf-8'))
    items = response_dict['body']['items']

    df = pd.DataFrame(items)
    return df
```

```python
skey = 'R0Wnbv0TpQW2w3AH8ezK89SSX3e1rFBgUq4lhFuA3oek1BOSEFlSvQSJ0mxXQ7NK%2Bq5rNWSy7VJKU0VQykvEUg%3D%3D'
res_id = 'storezone'
encode = urlencode({ quote_plus('key') : '9174' })
baro_api(skey, res_id, encode)
```
